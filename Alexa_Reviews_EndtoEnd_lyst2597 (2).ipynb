# NLP Preprocessing Pipeline


![](https://drive.google.com/uc?id=1kusknPPZ-2ur09oAaN0WHJDN3qVQDX5s)

```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_sm')
```

```python
text = df_reviews.loc[2,"verified_reviews"]
print(text)
```

### Parts of Speech and Lemmatization

```python
def pos(text):
  doc = nlp(text)
  print("TOKENIZATION --- POS TAG --- LEMATIZATION")
  for token in doc:
    print(token.text,'---', token.pos_,'---', token.lemma_)


pos(text)
```

### Remove Stopwords

```python
def stops(text):
  doc = nlp(text)
  cleaned = []
  for token in doc:
    if token.is_stop:
      continue
    else:
      cleaned.append(token.text)
  return cleaned

print("AFTER REMOVING STOPWORDS")
" ".join(stops(text))
```

### Dependency Parsing and Noun Chunks

```python
def noun_chunk(text):
  doc = nlp(text)
  print("HERE IS THE DEPENDENCY TREE")
  displacy.render(doc, style='dep', jupyter=True)
  print("HERE ARE THE NOUN CHUNKS")
  for chunk in doc.noun_chunks:
    print(chunk.text, "---", chunk.root.text)

noun_chunk(text)
```

### Named Entity Recognition

```python
def ner(text):
  doc = nlp(text)
  print("HERE ARE ALL THE NAMED ENTITIES")
  displacy.render(doc, style='ent', jupyter=True)
  
ner(text)
```

### Summary

```python
def text_cleaning(text):
  pos(text)
  stops(text)
  noun_chunk(text)
  ner(text)

text_cleaning(text)
```

## Text Classification using Elmo (Requires the use of CPU)

```python
df_reviews.head()
```

```python
data = df_reviews[['verified_reviews', 'feedback']]
```

```python
data = data[['feedback', 'verified_reviews']].rename(columns={"v1":"feedback", "v2":"verified_reviews"})
 
data['feedback'] = '__label__' + data['feedback'].astype(str)

data.iloc[0:int(len(data)*0.8)].to_csv('train.csv', sep='\t', index = False, header = False, columns=['feedback', 'verified_reviews'])
data.iloc[int(len(data)*0.8):int(len(data)*0.9)].to_csv('test.csv', sep='\t', index = False, header = False, columns=['feedback', 'verified_reviews'])
data.iloc[int(len(data)*0.9):].to_csv('dev.csv', sep='\t', index = False, header = False, columns=['feedback', 'verified_reviews'])
```

```python
from flair.data_fetcher import NLPTaskDataFetcher
from flair.embeddings import WordEmbeddings, DocumentLSTMEmbeddings, ELMoEmbeddings
from flair.models import TextClassifier
from flair.trainers import ModelTrainer
from pathlib import Path


corpus = NLPTaskDataFetcher.load_classification_corpus(Path('./'), test_file='test.csv', dev_file='dev.csv', train_file='train.csv')
word_embeddings = [WordEmbeddings('glove'), ELMoEmbeddings()]
document_embeddings = DocumentLSTMEmbeddings(word_embeddings, hidden_size=512, reproject_words=True, reproject_words_dimension=256)


classifier = TextClassifier(document_embeddings, label_dictionary=corpus.make_label_dictionary(), multi_label=False)
trainer = ModelTrainer(classifier, corpus)
trainer.train('resources/elmo-model', max_epochs=10)
```

Model is saved in the resources folder as best-model.pt and final-model.pt

### Loading a saved model and making predictions on a single sentence.

```python
from flair.data import Sentence
from flair.models import TextClassifier

# load the model you trained
model = TextClassifier.load('resources/elmo-model/final-model.pt')

sentence = Sentence("Alexa has the ability to listen to me all the time. I dont like this feature.")

# predict tags and print
model.predict(sentence)

print(sentence.to_tagged_string())
print(sentence.labels)
```

```python
data.info()
```

```python
data_df = pd.DataFrame({
    'text': data['verified_reviews'],
    'label':data['feedback']
})

data_df.head()
```

```python
data_df.shape
```

```python
train_df = data_df[:3000]
eval_df = data_df[3000:]
```

```python
train_df.shape
```

```python
from simpletransformers.model import TransformerModel
```

### BERT

```python
# Create a TransformerModel
model = TransformerModel('bert', 'bert-base-cased', args={'reprocess_input_data': True, 'overwrite_output_dir': True})
```

```python
# Train the model
model.train_model(train_df)
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```python
result
```
